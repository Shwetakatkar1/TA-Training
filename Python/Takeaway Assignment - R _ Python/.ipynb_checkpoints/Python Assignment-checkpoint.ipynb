## Questions 1 - 6 Utilize the sales data set. 


```python
# Q1 Find least sales amount for each item

def least_sales(df):
    ls = df.groupby(["Item"])["Sale_amt"].min().reset_index()
    return ls

least_sales(sales_df)
```

```python
# Q2 compute total sales at each year X region
def sales_year_region(df):
    syr = df.groupby([df['OrderDate'].dt.year,df['Region'],df['Item']])['Sale_amt'].sum()
    return syr

sales_year_region(sales_df)
```

```python
# Q3 append column with no of days difference from present date to each order date

def days_diff(df):
    df['days_diff']=df['OrderDate'].apply(lambda x:  pd.to_datetime('today').date()-x.date())
    return df['days_diff']
days_diff(sales_df)
sales_df.head()
```

```python
# Q4 get dataframe with manager as first column and  salesman under them as lists in rows in second column.

def mgr_slsmn(df):
    newdf=pd.DataFrame({'Manager':df['Manager'].unique()})
    newdf['list_of_salesmen']=newdf.Manager.apply(lambda x: list(df[df['Manager']==x]['SalesMan']))
    return newdf

mgr_slsmn(sales_df)
```

```python
# Q5 For all regions find number of salesman and number of units

def slsmn_units(df):
    return df.groupby('Region')[['SalesMan','Units']].count().rename(columns={'SalesMan':'salesmen_count','Units':'units_count'})

slsmn_units(sales_df)
```

```python
# Q6 Find total sales as percentage for each manager
def sales_pct(df):
    q10 = (df.groupby('Manager').sum()['Sale_amt']*100/df['Sale_amt'].sum()).reset_index().rename(columns={'Sale_amt':'percent_sales'})
    return q10

sales_pct(sales_df)
```

## Questions 7 - 10 Utilize the imdb data set (duration is in seconds)

```python
imdb_df = pd.read_csv('imdb.csv', escapechar='\\')
```

```python
imdb_df.head()
```

```python
# Q7 get imdb rating for fifth movie of dataframe

def fifth_movie(df):
    return df['imdbRating'][4]

fifth_movie(imdb_df)
```

```python
# Q8 return titles of movies with shortest and longest run time

def movies(df):
    short,long = df[df['duration']==df['duration'].min()]['title'].to_string(), df[df['duration']==df['duration'].max()]['title'].to_string()
    return short, long

movies(imdb_df)
```

```python
# Q9 sort by two columns - release_date (earliest) and Imdb rating(highest to lowest)

def sort_df(df):
    return df.sort_values(by = ['year', 'imdbRating'], ascending = [True, False], na_position = 'last')


sort_df(imdb_df)

```

```python
# Q10 subset revenue more than 2 million and spent less than 1 million & duration between 30 mintues to 180 minutes

def subset_df(df):
    return df[(df['duration'] >30*60) & (df['duration'] < 180*60)]

subset_df(imdb_df)
```

## Questions 11 - 15 Utilize the diamonds data set.


```python
diamonds=pd.read_csv('diamonds.csv')
```

```python
# Q11 count the duplicate rows of diamonds DataFrame.

def dupl_rows(df):
    return df.duplicated().sum()
    
dupl_rows(diamonds)
```

```python
# Q12 droping those rows where any value in a row is missing in carat and cut columns

def drop_row(df):
    return df.dropna(subset=['carat','cut'],how='any')
drop_row(diamonds)
```

```python
# Q13 subset only numeric columns

def sub_numeric(df):
    return df.select_dtypes(include=np.number)

sub_numeric(diamonds)
```

```python
# Q14 compute volume as (x*y*z) when depth > 60 else 8

def volume(df):
    df['z'] = pd.to_numeric(df['z'], errors='coerce')
    df['volume']=df.apply(lambda x: (x['x'])*(x['y'])*(x['z']) if x['depth']>60 else 8 , axis=1)
    return df
volume(diamonds)
```
